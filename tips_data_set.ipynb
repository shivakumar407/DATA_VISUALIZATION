```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
import plotly.express as px
import tensorflow as tf
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
#barplot
#average tips for each day of the week.
plt.figure(figsize=(8, 5 ))
sns.barplot(data=tips,x='day',y='tip',hue='sex')
plt.title('average tips by day and gender')
plt.show()
```

```python
#boxplot
plt.figure(figsize=(8,5))
sns.boxplot(data=tips,x='day',y='total_bill',hue='time')

plt.title("box plot of total bill by day and time")
palette='size'
plt.show()
```

```python
#violin plot
# dennsity of tips across gender
plt.figure(figsize=(8,5))
sns.violinplot(data=tips,x='sex', y='tip',hue='smoker',split='true')
plt.title("violin plot of tips by gender and smoker status")
plt.show()
```
